```python
# Import Pandas and pathlib
import pandas as pd
from pathlib import Path
```

```python
#Importing Datasets
song_data = pd.read_csv('Datasets/song_data.csv')
lyric_data = pd.read_csv('Datasets/lyrics-data.csv')
artist_data = pd.read_csv('Datasets/artists-data.csv')
song_data.head(5)
```

```python
artist_data.info()
print('\nColumns in artist_data:',artist_data.columns)
```

```python
# Dropping uneccessary columns 
lyric_data = lyric_data.drop(['SLink'], axis = 1)

song_data = song_data.drop(['type', 'uri', 'track_href', 'analysis_url', 'artists', 'id'], axis = 1)

artist_data = artist_data.drop(['Genres', 'Artist'], axis = 1)

# Dropping rows with popularity = 0 as they are not accurate

song_data = song_data[song_data.popularity != 0]
artist_data = artist_data[artist_data.Popularity != 0]

lyric_data.info()
```

```python
song_data.info()
```

```python
artist_data.info()
```

```python
#Looking for null values in datasets
print('song_data:')
print(song_data.isnull().sum())
print('\n')
print('lyric_data:')
print(lyric_data.isnull().sum())
print('\n')
print('artist_data:')
print(artist_data.isnull().sum())
```

```python
#Dropping NULL Values
song_data = song_data.dropna()
print('song_data:')
print(song_data.count())
print('')
lyric_data = lyric_data.dropna()
print('lyric_data:')
print(lyric_data.count())
print('')
artist_data = artist_data.dropna()
print('artist_data:')
print(artist_data.count())
```

```python
#Looking for duplicates in song_data
duplicate_rows_df = song_data[song_data.duplicated()]
print('number of duplicate rows: ', duplicate_rows_df.shape)
```

```python
#Looking for duplicates in lyric_data
duplicate_rows_df = lyric_data[lyric_data.duplicated()]
print('number of duplicate rows: ', duplicate_rows_df.shape)
```

```python
#Looking for duplicates in artist_data
duplicate_rows_df = artist_data[artist_data.duplicated()]
print('number of duplicate rows: ', duplicate_rows_df.shape)
```

```python
#Dropping the duplicates
song_data = song_data.drop_duplicates()
print('song_data after removing duplicates:')
print(song_data.count())
print('')
lyric_data = lyric_data.drop_duplicates()
print('lyric_data after removing duplicates:')
print(lyric_data.count())
print('')
artist_data = artist_data.drop_duplicates()
print('artist_data after removing duplicates:')
print(artist_data.count())
```

```python
#Changing artist name in lyric data to match format in song_data
lyric_data['ALink'] = lyric_data['ALink'].str.replace('-',' ')
lyric_data['ALink'] = lyric_data['ALink'].str.replace('/','')
lyric_data.head(5)
```

```python
#Changing artist name in artist_data data to match format in song_data
artist_data['Link'] = artist_data['Link'].str.replace('-',' ')
artist_data['Link'] = artist_data['Link'].str.replace('/','')
artist_data.head(5)
```

```python
# Creating a temp df for song_data so that the song_name and artist_name match
temp_song = song_data
temp_song['name.1'] = temp_song['name.1'].str.lower()
temp_song.head(5)
```

## Merging the datasets

```python
#Using the artist name as the key for merging the datasets lyric_data and artist_data

lyric_data = lyric_data.merge(artist_data, left_on = ['ALink'], right_on = ['Link'],how='inner').drop(['Link'], axis = 1)
lyric_data.info()
```

```python
#Using the Song id as the key for merging the datasets

data = temp_song.merge(lyric_data, left_on = ['name.2','name.1'], right_on = ['SName','ALink'],how='inner').drop(['SName','ALink'], axis = 1)
data.info()
```

```python
data.head(5)
```

```python
data.info()
print('\nColumns in data:',data.columns)
```

```python
# Dropping non english songs

data = data.loc[data["language"] == 'en' ]
data.info()
```

```python
#Renaming Columns for better readabilty

data = data.rename(columns={"name": "album_name", "name.1": "artist_name", "name.2": "song_name", "id.1": "id", "Lyric" : "lyrics", "Popularity" : "artist_popularity", "Songs" : "artist_songs", "total_tracks" : "album_total_tracks"})
data.head(5)
```

```python
# Dropping duplicates with same song name
data = data.drop_duplicates(subset=['song_name'], keep='last')
data.head(5)
```

## Scaling the numerical values

```python
#making a new dataset with only numerical data
comp_data = data.select_dtypes(include=['float64','int64'])
comp_data.info() 

#making a new dataset with only non numeric data
nn_data = data[['album_name', 'release_date', 'artist_name','explicit', 'song_name','lyrics', 'language']]
```

```python
# Min-Max Normalization
comp_data = (comp_data-comp_data.min())/(comp_data.max()-comp_data.min())
data = pd.concat((nn_data,comp_data), 1)
 
print("Scaled Dataset Using Pandas")
data.head(5)
```

```python
# Reindexing
data = data.reset_index(drop=True)
data.head(5)
```

```python
data.info()
print('\nColumns in data:',data.columns)
```

### Making a new csv file with cleaned data

```python
# Cleaned dataframe is saved as a csv file
filepath = Path('Datasets/cleaned_data.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
data.to_csv(filepath)  
```
